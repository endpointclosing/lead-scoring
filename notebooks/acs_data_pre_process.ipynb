```python
states = {
    'AK': 'Alaska',
    'AL': 'Alabama',
    'AR': 'Arkansas',
    'AS': 'American Samoa',#
    'AZ': 'Arizona',
    'CA': 'California',
    'CO': 'Colorado',
    'CT': 'Connecticut',
    'DC': 'District of Columbia',
    'DE': 'Delaware',
    'FL': 'Florida',
    'GA': 'Georgia',
    'GU': 'Guam',#
    'HI': 'Hawaii',
    'IA': 'Iowa',
    'ID': 'Idaho',
    'IL': 'Illinois',
    'IN': 'Indiana',
    'KS': 'Kansas',
    'KY': 'Kentucky',
    'LA': 'Louisiana',
    'MA': 'Massachusetts',
    'MD': 'Maryland',
    'ME': 'Maine',
    'MI': 'Michigan',
    'MN': 'Minnesota',
    'MO': 'Missouri',
    'MP': 'Northern Mariana Islands',#
    'MS': 'Mississippi',
    'MT': 'Montana',
    'NC': 'North Carolina',
    'ND': 'North Dakota',
    'NE': 'Nebraska',
    'NH': 'New Hampshire',
    'NJ': 'New Jersey',
    'NM': 'New Mexico',
    'NV': 'Nevada',
    'NY': 'New York',
    'OH': 'Ohio',
    'OK': 'Oklahoma',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'PR': 'Puerto Rico',#
    'RI': 'Rhode Island',
    'SC': 'South Carolina',
    'SD': 'South Dakota',
    'TN': 'Tennessee',
    'TX': 'Texas',
    'UT': 'Utah',
    'VA': 'Virginia',
    'VI': 'Virgin Islands',#
    'VT': 'Vermont',
    'WA': 'Washington',
    'WI': 'Wisconsin',
    'WV': 'West Virginia',
    'WY': 'Wyoming'
}

def state_std(x:str):

    if x is None:
        return ""

    elif (len(x) == 2) & (x.isalpha()):
        return x.upper()
    else:
        for key, value in states.items():
            if x.lower() == value.lower():
                return key
        return ""
```

```python
state_acs_dp02 = pd.read_csv("../data/census/state_level/ACSDP5Y2020.DP02_data_with_overlays_2022-06-02T154134.csv", header=1)
state_acs_dp03 = pd.read_csv("../data/census/state_level/ACSDP5Y2020.DP03_data_with_overlays_2022-06-02T154134.csv", header=1)
state_acs_dp04 = pd.read_csv("../data/census/state_level/ACSDP5Y2020.DP04_data_with_overlays_2022-06-02T154134.csv", header=1)
state_acs_dp05 = pd.read_csv("../data/census/state_level/ACSDP5Y2020.DP05_data_with_overlays_2022-06-02T154134.csv", header=1)
```

```python
state_acs = pd.DataFrame()
for df in [state_acs_dp02,state_acs_dp03,state_acs_dp04, state_acs_dp05]:
    df = df[[i for i in df.columns if (i.startswith("Percent!!") & all(df[i] != "(X)"))] + ["Geographic Area Name"]]
    if state_acs.shape[0] == 0:
        state_acs = df
    else:
        state_acs = state_acs.merge(df, on = "Geographic Area Name")
```

```python
state_acs = state_acs[["Geographic Area Name"] +state_acs.columns.drop("Geographic Area Name").tolist()]
```

```python
state_acs["STATE_STD"] = state_acs["Geographic Area Name"].apply(state_std)
```
